```python
# importa as bibliotecas
import gurobipy as gp
from gurobipy import GRB
```

```python
# criando o modelo
m = gp.Model("moveis")
```

```python
# criando as variaveis
#x1 refere-se a cadeiras
#x2 refere-se a mesas
x1 = m.addVar(name="x1")
x2 = m.addVar(name="x2")
#x1 = m.addVar(vtype=GRB.INTEGER,name="x1")
#x2 = m.addVar(vtype=GRB.INTEGER,name="x2")
```

```python
# definindo função objetivo
m.setObjective(45*x1 + 80*x2,GRB.MAXIMIZE)
```

```python
# adicionar as restrições
con1 = m.addConstr(5*x1 + 20*x2 <= 400, "madeira")
con2 = m.addConstr(10*x1 + 15*x2 <= 450, "mao-de-obra")
```

```python
# salvando modelo
m.write("moveis.lp")
```

```python
m.optimize()
```

```python
for v in m.getVars():
    print(v.varName,"=", v.x)
```
