```python
# importar as bibliotecas
import gurobipy as gp
from gurobipy import GRB
```

Considere o seguinte MIP.

$
\begin{align*}
\max \ & x +   y + 2 z \\
& x + 2 y + 3 z \leq 4 \\
& x +   y      \geq 1 \\
& x, y, z \in \mathbb{B} \\
\end{align*}
$

Implemente o modelo usando API gurobipy.

```python
# criando o modelo
modelo = gp.Model()
```

```python
# adicionando as variaveis
x = modelo.addVar(vtype = GRB.BINARY, name = "x")
y = modelo.addVar(vtype = GRB.BINARY, name = "y")
z = modelo.addVar(vtype = GRB.BINARY, name = "z")
```

```python
# definindo a funcao objetivo
modelo.setObjective(x + y + 2*z, GRB.MAXIMIZE)
```

```python
# adicionando as restrições
rest1 = modelo.addConstr(x + 2*y + 3*z <= 4, name = "rest1")

rest2 = modelo.addConstr(x + y >= 1, name = "rest2")
```

```python
# exportando .lp/.mps
modelo.write("binary.mps")
```

```python
# resolvendo o problema
modelo.optimize()
```

```python
print("Valor máximo = ", modelo.objVal)
```

```python
# imprime solução
for i in modelo.getVars():
    print(i.varName, "=", i.x)
```
